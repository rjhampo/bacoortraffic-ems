```python
import json
import pandas as pd
```

```python
with open('C:/Users/James Ampo/Documents/jupyter/sparta_capstone/GIS/hospital_pois.geojson', 'r') as file:
    pois = file.read()
    file.close()
```

```python
pois = json.loads(pois)
pois
```

```python
names = []
is_public = []
x_coords = []
y_coords = []
for feature in pois['features']:
    names.append(feature['properties']['name'] or feature['properties']['hospname'])
    is_public.append(feature['properties']['public'])
    x_coords.append(feature['geometry']['coordinates'][0])
    y_coords.append(feature['geometry']['coordinates'][1])
```

```python
for i in range(len(names)):
    print(i, names[i], is_public[i])
```

```python
# Manually modifying is_public values for last 3 rows
# As last 3 rows are combined hospitals which some are public
# if there is one public hospital, the point is assigned 1

is_public[31] = 1
is_public[32] = 1
is_public[33] = 0
```

```python
hosp_df = pd.DataFrame([names, is_public, x_coords, y_coords])
hosp_df = hosp_df.transpose()
hosp_df.columns = ['hospital_name', 'is_public', 'x_coord', 'y_coord']
```

```python
hosp_df.to_csv('hospital_coords.csv', index=False)
```
